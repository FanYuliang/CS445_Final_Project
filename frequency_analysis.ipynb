```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
im1_file = './resampled.png'
```

```python

im1 = cv2.imread(im1_file, cv2.IMREAD_GRAYSCALE)
```

```python


fig, axes = plt.subplots(1, 2,figsize=(9, 5))
axes[0].imshow(im1, cmap='gray')
axes[0].set_title('Unsmoothed Image'), axes[0].set_xticks([]), axes[0].set_yticks([])
axes[1].imshow(np.log(np.abs(np.fft.fftshift(np.fft.fft2(im1)))), cmap='gray')
axes[1].set_title('Frequency Domain'), axes[1].set_xticks([]), axes[1].set_yticks([])

plt.show()
plt.savefig('frequency_domain.png')
```
